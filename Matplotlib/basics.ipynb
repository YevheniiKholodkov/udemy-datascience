```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = x * 2
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0, 6)
plt.ylim(0, 15)
plt.title('String title')
plt.savefig('my_plot.png', dpi=300)
```
