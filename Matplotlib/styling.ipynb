```python
x = np.arange(0, 10)
y = x * 2 
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X^2')
ax.legend(loc=(1.2, 0.5))
fig.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='#e6c017', lw=0.5, marker='+')
lines = ax.plot(x, x+1, color='purple', linewidth=2, markerfacecolor='red', marker='o', linestyle='--')
lines[0].set_dashes([1, 2, 3, 5])
```
